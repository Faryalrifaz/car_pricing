# **Visualization**

Visualizations play a key role in data analysis. In this reading, you'll be introduced to various forms of graphs and plots that you can create with your data in Python that help you in visualising your data for better analysis.

The two major libraries used to create plots are matplotlib and seaborn. We will learn the prominent plotting functions of both these libraries as applicable to Data Analysis.

## Import Libraries 

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

## Import Data

```python
df = pd.read_csv("F:\\projects\\car_pricing\\data\\clean_df.csv")
X = df.drop('price', axis=1)
y = df['price']
print(df.head)
```

## Display the data in table

1. Standard Line Plot

The simplest and most fundamental plot is a standard line plot. The function expects two arrays as input, x and y, both of the same size. x is treated as an independent variable and y as the dependent one. The graph is plotted as shortest line segments joining the x,y point pairs ordered in terms of the variable x.


```python
plt.plot(X['wheel-base'], y, label='Wheel Base vs Price')
plt.xlabel('Wheel Base')
plt.ylabel('Price')
plt.title('Standard Line Plot')
plt.legend()
plt.show()
```

## Scatter Plot

```python
plt.scatter(X['curb-weight'], y, alpha=0.7)
plt.xlabel('Curb Weight')
plt.ylabel('Price')
plt.title('Price vs Curb Weight')
plt.show()
```

```python
## Correlation heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
## Pairplot for selected features
selected_features = ['wheel-base', 'curb-weight', 'engine-size', 'horsepower', 'price']
sns.pairplot(df[selected_features], diag_kind='kde', corner=True)
plt.suptitle('Pairplot of Selected Features', y=1.02)
plt.show()
```

```python
## Distribution of price
plt.figure(figsize=(10, 6))
sns.histplot(y, kde=True, bins=30, color='blue')
plt.title('Distribution of Price')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

##  Boxplot for price vs engine size

```python

plt.figure(figsize=(12, 6))
sns.boxplot(x=pd.cut(df['engine-size'], bins=5), y=y)
plt.title('Boxplot of Price vs Engine Size')
plt.xlabel('Engine Size (Binned)')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.show()
```
